```python
# dataset choose
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X=cancer.data
Y=cancer.target
df=pd.DataFrame(X,columns=cancer.feature_names)
df['Target']=Y
df
```

```python
#Cleaning data by handling missing values and removing duplicates
print(df.isnull().sum())
print(df.duplicated().sum())
#Converting categorical data using encoding techniques
#Target variable is already encoded so no need to encode again
#Features variables standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=df.drop('Target',axis=1)
X_scaled=scaler.fit_transform(X)
print(X_scaled)
```

```python
#Splitting the dataset into training and testing sets

Y=df['Target']
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.4,random_state=42)
```

perform an exploratory data analysis to visualize and summarize the data:

Distribution of classes
our Target column representing the output variable is constituted of two classes:
Begnin tumor=0
Malignant tumor =1

```python
#Correlation between features
correlation_matrix=df.corr()
correlation_matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,18))
sns.heatmap(correlation_matrix,annot=True,cmap='viridis',linewidths=0.5,fmt='.2f')
plt.show()
```

```python
#Model Implementation and Evaluation:

#Implement at least two classification models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report

#Decision Tree

dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
y_pred_dt=dt_model.predict(x_test)
accuracy_dt=accuracy_score(y_test,y_pred_dt)
classification_report_dt=classification_report(y_test,y_pred_dt)
print("Decision Tree Accuracy:",accuracy_dt)
print("Decision Tree Classification Report:\n",classification_report_dt)

#Random Forest
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
y_pred_rf=rf_model.predict(x_test)
accuracy_rf=accuracy_score(y_test,y_pred_rf)
classification_report_rf=classification_report(y_test,y_pred_rf)
print("Random Forest Accuracy:",accuracy_rf)
print("Random Forest Classification Report:\n",classification_report_rf)
```

Potential Issues During Deployment & Operation
1. Data Drift
In production, the data the model receives may slowly change over time (new patient profiles, new measurement patterns).
This can reduce accuracy.
How to handle it:
- Continuously monitor feature distributions
- Use drift‑detection tools (Evidently AI, MLflow)
- Retrain the model periodically with fresh data
2. Real‑Time Data Quality Problems
Incoming real‑time data may be:
- Missing values
- Incorrectly formatted
- Out of expected ranges
- Delayed or duplicated
How to handle it:
- by adding validation rules before prediction
- by rejecting or correcting invalid inputs
- by Login all anomalies for later review

3. Model Versioning & Rollback
If we update the model, a new version may perform worse in production.
How to handle it:
- by keeping all versions stored
- by using  safe deployment strategies (canary, blue‑green)
- by allowing instant rollback to the previous version

4. Security & Privacy Risks
Medical data (like breast cancer features) is sensitive.
How to handle it:
- by encrypting data in transit (HTTPS)
- by restricting API access (tokens, authentication)
- by Login predictions without storing personal identifiers




1. Monitor Model Performance Over Time
A model that performs well at deployment may degrade as real‑world data evolves.
What to track:
- Accuracy, Precision, Recall, F1
- Confusion matrix trends
- False positives / false negatives (critical in medical contexts)
How to maintain:
- Set performance thresholds
- Trigger alerts when metrics drop
- Schedule periodic evaluations (weekly or monthly)
 2. Detect Data Drift
Real‑world data may slowly change (new patient demographics, new measurement devices, etc.).
Types of drift:
- Feature drift: input distributions change
- Prediction drift: output distribution changes
- Concept drift: the relationship between features and labels changes
How to maintain:
- Use drift‑detection tools (Evidently AI, MLflow)
- Compare live data distributions to training data
- Retrain the model when drift exceeds a threshold

 3. Monitor Data Quality in Real Time
Incoming data may be incomplete, noisy, or invalid.
What to check:
- Missing values
- Out‑of‑range values
- Wrong data types
- Unexpected feature formats
How to maintain:
